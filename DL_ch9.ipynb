# CT스캔이란

단일 채널의 3차원 배열로 표현되는 3차원 엑스레이.
CT스캔 데이터는 흑백 PNG 이미지를 차곡차곡 쌓아논것과 같다. 엑스레이의 경우 3차원적인 강도를 2차원 평면에 투영한것이지만 CT스캔 데이터는 3차원 형태를 보존함

- 복셀(voxel): 2차원 픽셀의 3차원 버전 

의학 데이터 외에도 유체 시뮬레이션, 2차원 이미지로 부터 구성된 3차원 장면, 자율주행자동차의 라이다 데이터등에서 복셀데이터를 볼 수 있다.

# 엔드투엔드 폐암 진단기
프로젝트 구성: 디스크의 대부분은 밀도정보를 가진 CT스캔의 3차원 배열을 저장, 모델은 3차원 배열부분에 해당하는 여러 단면사용

엔드투엔드 솔루션은 CT데이터 파일 읽어서 CT인스턴스를 만든 후 전체 3차원 스캔 데이터를 담고 세그멘테이션(관심있는 복셀 표시)을 수행하는 모듈과 결합한 후 복셀을 작은 덩어리로 나누어 결절로 간주될만한것을 찾음.

1. 로딩

2. 세그멘테이션

3. 그룹화

4. 분류

5. 결절분석 및 진단

# 결절 설명

장기 내부에 나타날수있는 돌기나 덩어리.
암은 항상 결절의 형태니까 이런 결절에만 집중하면 된다.
